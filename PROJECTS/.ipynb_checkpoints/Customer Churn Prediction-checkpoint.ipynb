# Outlier Detection And Removal

```python
# Outlier Detection Using Boxplot Method
numeric_cols = df.select_dtypes(include=['float64','int64']).columns
numeric_cols = numeric_cols.tolist()
numeric_cols.remove("Churn")
print(numeric_cols)
```

```python
fig , axes = plt.subplots(2,2 , figsize=(12,6))

# Flatten axes if needed
axes = axes.flatten()

# Iterate over numeric columns and create box plots in each subplot
for i, column in enumerate(numeric_cols):
    df[column].plot(kind='box', ax=axes[i])
    axes[i].set_title(f'Box Plot - {column}')
    
plt.tight_layout()
plt.show()
```

```python
# Outlier Detection Using z-score Method

# Filter numeric columns
numeric_columns = df.select_dtypes(include=['float64', 'int64']).columns

# Set Z-score threshold 
z_score_threshold = 2.5

# Detect outliers for each numeric column
outliers = (np.abs(stats.zscore(df[numeric_columns])) > z_score_threshold)

# Get numeric columns with outliers
numeric_columns_with_outliers = numeric_columns[outliers.any()]

# Display numeric columns with outliers
print(numeric_columns_with_outliers)
```

**By performing Exploratory Data Analysis and Outlier Detection We found that Every column follows same kind of Distribution for Both Chrun and Not Churn and No one contains outlier.**

# Feature Engineering
** **
1. Encoding the Labels
2. Creating features from Existing feautres
3. Feature Scaling

```python
# Encoding the labels
encoder = LabelEncoder()
```

```python
for i in df[['Gender','Location']]:
    df[i] = encoder.fit_transform(df[i])
```

```python
df.info()
```

```python
# Generating new Features using Data Descretization
df['New_Age'] = pd.qcut(df['Age'] , 3 , labels=[1,2,3]).astype('int64')
df['New_Monthly_Bill'] = pd.qcut(df['Monthly_Bill'] , 5 , labels = [1,2,3,4,5]).astype('int64')
df['New_Total_Usage_GB'] = pd.qcut(df['Total_Usage_GB'] , 6 , labels = [1,2,3,4,5,6]).astype('int64')
```

```python
df.head()
```

```python
# Generating more Features using Ratio Method
df['Age_Subscription'] = df['Age'] / df['Subscription_Length_Months']
df['Total_Usage_Montly_Bill'] = df['Total_Usage_GB'] / df['Monthly_Bill']
df['Usage_per_Month'] = df['Subscription_Length_Months'] / df['Total_Usage_GB']
df['Total_Usage_Subscription'] = df['Total_Usage_GB'] / df['Subscription_Length_Months']
```
